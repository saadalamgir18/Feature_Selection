```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('train.csv')
```

```python
X = df.drop("SalePrice",axis=1)   #Feature Matrix
y = df["SalePrice"] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

```python
num_features = [feature for feature in X_train.columns if X_train[feature].dtype != 'O']
```

```python
import seaborn as sns
#Using Pearson Correlation
plt.figure(figsize=(12,10))
cor = X_train[num_features].iloc[:,:10].corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train[num_features], 0.7)
len(set(corr_features))
```
